```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# MLM vs GLM

```python
def draw_pretrain_diff(df):
    g = sns.catplot(
        data=df, kind="bar",
        x="lan", y="score", hue="procedure",
        ci="sd", palette="dark", alpha=.6, height=6,
        hue_order=['MLM', 'GLM']
    )
    g.despine(left=True)
    g.set_axis_labels("language", "Test score")
    g.legend.set_title("")
```

```python
xlmr_base_df = pd.DataFrame([
    ['GLM', 'en', 85.1],
    ['GLM', 'ru', 80.1],
    ['GLM', 'fr', 80.5],
    ['GLM', 'ar', 79.2],
    ['GLM', 'zh', 79.1],
    
    ['MLM', 'en', 78.8],
    ['MLM', 'ru', 63.6],
    ['MLM', 'fr', 69.6],
    ['MLM', 'ar', 67.1],
    ['MLM', 'zh', 72.1],
], columns=['procedure', 'lan', 'score'])

draw_pretrain_diff(xlmr_base_df)
plt.title('XLM-Roberta base');
```

```python
xlmr_large_df = pd.DataFrame([
    ['GLM', 'en', 89.5],
    ['GLM', 'ru', 85.7],
    ['GLM', 'fr', 85.7],
    ['GLM', 'ar', 82.6],
    ['GLM', 'zh', 83.5],
    
    ['MLM', 'en', 72.1],
    ['MLM', 'ru', 61.1],
    ['MLM', 'fr', 62.3],
    ['MLM', 'ar', 64.5],
    ['MLM', 'zh', 71.1],
], columns=['procedure', 'lan', 'score'])

draw_pretrain_diff(xlmr_large_df)
plt.title('XLM-Roberta large');
```

```python
bert_base_df = pd.DataFrame([
    ['GLM', 'en', 88],
    ['MLM', 'en', 82.5],
], columns=['procedure', 'lan', 'score'])

draw_pretrain_diff(bert_base_df)
plt.title('BERT base');
```

# Dist diffs

```python
def draw_method_diff(df):
    g = sns.catplot(
        data=df, kind="bar",
        x="lan", y="score", hue="method",
        ci="sd", palette="dark", alpha=.6, height=6
    )
    g.despine(left=True)
    g.set_axis_labels("language", "Test score")
    g.legend.set_title("")
```

```python
glm_xlmr_base_df = pd.DataFrame([
    ['Dot Probs', 'en', 76.1],
    ['Jen-Shen Probs', 'en', 77.5],
    ['Dot Embs, p=2', 'en', 84.3],
    ['Dist Embs, p=2', 'en', 83.6],
    ['Dist Embs, p=1', 'en', 85.1],
    
    ['Dot Embs, p=2', 'ru', 79.8],
    ['Dist Embs, p=2', 'ru', 79.2],
    ['Dist Embs, p=1', 'ru', 80.1],
    
    ['Dot Embs, p=2', 'fr', 78],
    ['Dist Embs, p=2', 'fr', 78.2],
    ['Dist Embs, p=1', 'fr', 80.5],
    
    ['Dot Embs, p=2', 'ar', 77],
    ['Dist Embs, p=2', 'ar', 76.9],
    ['Dist Embs, p=1', 'ar', 79.2],
    
    ['Dot Embs, p=2', 'zh', 77.9],
    ['Dist Embs, p=2', 'zh', 77.6],
    ['Dist Embs, p=1', 'zh', 79.1],
], columns=['method', 'lan', 'score'])

draw_method_diff(glm_xlmr_base_df)
plt.title('GLM XLM-Roberta base');
```

```python
glm_xlmr_large_df = pd.DataFrame([
    ['Dot Probs', 'en', 81.5],
    ['Jen-Shen Probs', 'en', 83.8],
    ['Dot Embs, p=2', 'en', 86.4],
    ['Dist Embs, p=2', 'en', 86.8],
    ['Dist Embs, p=1', 'en', 89.5],
    
    ['Dot Embs, p=2', 'ru', 82.2],
    ['Dist Embs, p=2', 'ru', 82.2],
    ['Dist Embs, p=1', 'ru', 85.7],
    
    ['Dot Embs, p=2', 'fr', 83.8],
    ['Dist Embs, p=2', 'fr', 84.2],
    ['Dist Embs, p=1', 'fr', 85.7],
    
    ['Dot Embs, p=2', 'ar', 80.6],
    ['Dist Embs, p=2', 'ar', 79.9],
    ['Dist Embs, p=1', 'ar', 82.6],
    
    ['Dot Embs, p=2', 'zh', 79.7],
    ['Dist Embs, p=2', 'zh', 80.5],
    ['Dist Embs, p=1', 'zh', 83.5],
], columns=['method', 'lan', 'score'])

draw_method_diff(glm_xlmr_large_df)
plt.title('GLM XLM-Roberta large');
```

```python
glm_bert_base_df = pd.DataFrame([
    ['Dot Probs', 'en', 79.8],
    ['Jen-Shen Probs', 'en', 84],
    ['Dot Embs, p=2', 'en', 87.1],
    ['Dist Embs, p=2', 'en', 86.9],
    ['Dist Embs, p=1', 'en', 88],
], columns=['method', 'lan', 'score'])

draw_method_diff(glm_bert_base_df)
plt.title('GLM BERT base');
```
